# Location information extracting

```python
import math

def encode_geohash(latitude, longitude, precision):
    # Set up precision variables
    precision = precision * 5 // 2
    min_latitude, max_latitude = -90, 90
    min_longitude, max_longitude = -180, 180
    
    # Set up bit masks
    bit_masks = [16, 8, 4, 2, 1]
    
    # Create geohash
    geohash = ""
    is_even = True
    current_bit, bit_mask = 0, bit_masks[0]
    while len(geohash) < precision:
        if is_even:
            mid = (min_longitude + max_longitude) / 2
            if longitude > mid:
                current_bit |= bit_mask
                min_longitude = mid
            else:
                max_longitude = mid
        else:
            mid = (min_latitude + max_latitude) / 2
            if latitude > mid:
                current_bit |= bit_mask
                min_latitude = mid
            else:
                max_latitude = mid
        
        is_even = not is_even
        
        if bit_mask == 1:
            geohash += "0123456789bcdefghjkmnpqrstuvwxyz"[current_bit]
            current_bit, bit_mask = 0, bit_masks[0]
        else:
            bit_mask = bit_masks[bit_masks.index(bit_mask) + 1]
    
    return geohash
```

```python
# Create a new column called geohash and encode the latitude and longitude into it
df['geohash'] = df.apply(lambda x: encode_geohash(x['Latitude'], int(x['Longitude']),3), axis=1)

# Show the data
print(df.head())
```

# LSTM

```python
# define x as df
X = df
# define y as target longitudes, latitudes and time difference
y = df[['Time difference','geohash']]
```

```python
np.set_printoptions(suppress=True)
```

```python
def create_time_windows(x, y, window_size, step_size):
    X = []
    y_new = []
    for i in range(0, len(x) - window_size, step_size):
        X.append(x[i:i + window_size])
        y_new.append(y.iloc[i + window_size])
    return np.array(X), np.array(y_new)
```

```python
window_size = 7
step_size = 1

X, y_new = create_time_windows(X, y, window_size, step_size)
```

```python
# check both X and y_new shapes
print(X.shape)
print(y_new.shape)
```

```python
y_new_3d = y_new.reshape(y_new.shape[0], 1, y_new.shape[1])
print(y_new_3d.shape)
```

```python
from sklearn.model_selection import train_test_split

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_new_3d, test_size=0.2, random_state=42)
```

```python
# print x_train shape
print(X_train.shape)
# print x_test shape
print(X_test.shape)
# print y_train shape
print(y_train.shape)
# print y_test shape
print(y_test.shape)
```

```python
# import libraries for the model
from keras.models import Sequential
from keras.layers import Dense, LSTM , Flatten
# import softmax activation function
from keras.layers import Activation
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor

import numpy as np

from keras.callbacks import EarlyStopping

# Define the LSTM model
def create_model(num_layers, num_hidden_units):
    model = Sequential()
    for i in range(num_layers):
        if i == 0:
            # First layer requires input_shape argument
            model.add(LSTM(num_hidden_units[i], return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
        else:
            model.add(LSTM(num_hidden_units[i], return_sequences=True))

    model.add(Flatten())
    model.add(Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mse'])
    return model
```

```python

# Define the hyperparameters to search over
param_grid = {'num_layers': [1, 2],
              'num_hidden_units': [(32,), (64,)],
              'batch_size': [1, 16, 32, 64],
              'epochs': [10, 20],
              }

# Create an instance of the EarlyStopping callback
early_stopping = EarlyStopping(patience=5)

# Create the LSTM model using KerasRegressor for compatibility with scikit-learn
model = KerasRegressor(build_fn=create_model,validation_split=0.2)

# Perform grid search with early stopping
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(X_train, y_train, callbacks=[early_stopping])

# Print the best hyperparameters and mean absolute error score
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# create the model with the best parameters
model = Sequential()
model.add(LSTM(32, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Flatten())
model.add(Dense(2))
model.compile(optimizer='rmsprop', loss='mae', metrics=['mse'])
model.summary()
```

```python
# fit the model
history = model.fit(X_train, y_train, epochs=10, batch_size=16, validation_split=0.2, verbose=1)
```

```python
# make predictions
y_pred = model.predict(X_test)

# show the first 5 predictions
print("predicted results:", y_pred[:5])
```
